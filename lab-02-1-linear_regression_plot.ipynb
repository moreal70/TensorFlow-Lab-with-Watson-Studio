## Lab 2-1 Linear Regression with plot
* ***여기서는 그래프를 그려가면서 linear regression 이 어떻게 작동하는 지를 이해합니다.***
* ***아래쪽에 matplotlib 예시 코드를 참조해서 그래프 모양을 바꾸어 봅니다.***
* ***그래프 변수들도 다양하게 바꿔 가면서 돌려 봅니다.***

```python
# Lab 2 Linear Regression
import tensorflow as tf
tf.set_random_seed(777)  # for reproducibility
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
# X and Y data
x_train = [1, 2, 3]
y_train = [0.5, 1, 1.5]

# Try to find values for W and b to compute y_data = x_data * W + b
# We know that W should be 1 and b should be 0
# But let TensorFlow figure it out
W = tf.Variable(tf.random_normal([1]), name='weight')
b = tf.Variable(tf.random_normal([1]), name='bias')


# Our hypothesis XW+b
hypothesis = x_train * W + b

# cost/loss function
cost = tf.reduce_mean(tf.square(hypothesis - y_train))

# Minimize
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.01)
train = optimizer.minimize(cost)

# Launch the graph in a session.
sess = tf.Session()

# 그래프를 그리기 위한 변수들을 정의합니다.
x = [] 
y1 = []
y2 = []
y3 = []
y4 = []

# Initializes global variables in the graph.
sess.run(tf.global_variables_initializer())
```

```python
for step in range(2500):
    _ , W_val,cost_val,b_val,hypor_val = sess.run([train,W,cost,b,hypothesis])
    x.append(step)
    y1.append(W_val)
    y2.append(cost_val)
    y3.append(b_val)
    y4.append(hypor_val)
    if step % 50 == 0:
        print(step, sess.run(cost), sess.run(W), sess.run(b))
#plt.plot(x,y1,c='r',label='W')
#plt.plot(x,y2,c='b',label='cost')
#plt.plot(x,y3,c='y',label='b')
#plt.plot(x,y4,c='y',label='b')
plt.plot(y1,y2,c='y',label='w-cost')
plt.legend(shadow=True,fancybox=True,loc='upper right')
plt.ylim(0,1)
plt.show()   
```
